```python
import numpy as np
```

```python
# Generate transcation details
transction = np.random.randint(-100000, 100000, (50, 20))
```

```python
# Seperate debit and credit transction
debit_transction = transction[np.where(transction < 0)]
debit_transction
```

```python
# Seperate credit transction
credit_transction = transction[np.where(transction >= 0)]
credit_transction
```

```python
# Daily net balance change
np.sum(transction, axis=0)
```

```python
# Identify account with continues negative balance
cum_sum = np.cumsum(transction, axis=0)
cum_sum
```

```python
# Modify array to test
cum_sum[1] = [-1, -23, -367, -345, -46, -84, -65, -66, -97, -98, -1, -23, -367, -345, -46, -84, -65, -66, -97, -98]

# Accounts having continuous negative balance
negative_balance = np.all(cum_sum < 0, axis=1)
np.where(negative_balance)[0]
```
