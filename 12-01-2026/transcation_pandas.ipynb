```python
import pandas as pd
```

```python
# Read data from file
transction = pd.read_csv('transaction_log.csv')
transction
```

```python
# Average daily transction
transction.groupby('account_id')['amount'].mean()
```

```python
# Average number of transactions per account
avg_txn_count = transction.groupby('account_id')['amount'].count().mean()

# Accounts with higher than average transaction volume
transction.groupby('account_id')['amount'].count().loc[lambda x: x > avg_txn_count]
```

```python
# Detect potential fraud using statistical thresholds
txn_mean = transction['amount'].mean()
txn_std = transction['amount'].std()
threshhold_amount = txn_mean + 3 * txn_std
transction[transction['amount'] > threshhold_amount]
```
